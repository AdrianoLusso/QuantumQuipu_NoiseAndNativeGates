### First cost function

   Create a cost function for $Q$ number of qubits. The circuit of the cost function must be $I$ iterations of pairs of $rz$ gates. That means, if $I=3$, there sould be $2*3=6$ $rz$ gates per qubit. It also need a noise model as parameter. This noise model will affect the result of running the $rz$'s circuit. The cost is calculated as the sum of each qubit average result. Considering that we run the circuit $10000$, if we measure $6000$ times $0$ an the other $4000$ we measure $1$ in the first qubit, this qubit will have an average value of $0.6$. We repeat this process for each of the rest qubits and calculate the sum. The cost will be value between $0$ and $Q$

```python
def cost_func(num_qubits,num_iterations,noise_model):
    qc = QuantumCircuit(num_qubits,num_qubits)
    shots=10000

    #Add random phi cz gates
    for itr in range(num_iterations):
        random_phi = randrange(0,20,step=1)
        random_phi = random_phi / 10
        for q in range(num_qubits):
            qc.rz(random_phi * pi,q)
            qc.rz(random_phi * pi,q)

    qc.measure(range(num_qubits),range(num_qubits))
    display(qc.draw('mpl'))

    #Run the circuit 10000 times and store the results
    simulator = AerSimulator(noise_model=noise_model)
    result = simulator.run(qc,shots=shots).result()
    counts = result.get_counts(0)
    display(plot_histogram(counts,figsize=[18,7]))

    #Calculate the average for result of each qubit, and finally sum each of the averages to get the COST.
    cost = 0
    for qubit in range(num_qubits):
        qubit_average = 0
        for string in counts.keys():
            current_qubit_in_current_string = string[qubit]
            if current_qubit_in_current_string == '1':
                qubit_average += counts[string]
        qubit_average = qubit_average / shots
        print('qubit ',qubit,' average:',qubit_average,'\n')
        cost += qubit_average
    return cost

unm = Unified_Noise_Model()
#noise model with 0.2 of depolarization prob
unm.add_depolarizing_channel(0.1,['rz'],[])
print(unm.noise_model)

#cost function with 5 qubits and 4 iterations
cost = cost_func(4,4,unm.noise_model)
print('cost:',cost,'\n')
```

### Second cost function

Given a cost function with a fixed...
* amount of rz pairs iterations
* noise level
* amount of qubits

...find the rz rotations that minimize the cost.

* First of all, do it just with depolarizing noise.

```python
def cost_func2(rotations,num_qubits,num_iterations,noise_model):
    qc = QuantumCircuit(num_qubits,num_qubits)
    shots=10000

    #Add random phi cz gates
    for itr in range(num_iterations):
        for q in range(num_qubits):
            qc.rz(rotations[itr],q)
            qc.rz(rotations[itr],q)

    qc.measure(range(num_qubits),range(num_qubits))
    display(qc.draw('mpl'))

    #Run the circuit 10000 times and store the results
    simulator = AerSimulator(noise_model=noise_model)
    result = simulator.run(qc,shots=shots).result()
    counts = result.get_counts(0)
    #display(plot_histogram(counts,figsize=[18,7]))

    #Calculate the average for result of each qubit, and finally sum each of the averages to get the COST.
    cost = 0
    for qubit in range(num_qubits):
        qubit_average = 0
        for string in counts.keys():
            current_qubit_in_current_string = string[qubit]
            if current_qubit_in_current_string == '1':
                qubit_average += counts[string]
        qubit_average = qubit_average / shots
        #print('qubit ',qubit,' average:',qubit_average,'\n')
        cost += qubit_average
    cost = cost
    #print('cost ',cost)
    #print('---------------------------------------------')
    costs.append(cost)
    return cost
```

```python
from scipy.optimize import minimize
costs = []

# setting up the fixed parameters
depol_noise_prob = 0.5
num_iterations = 4
num_qubits = 4

#noise model
unm = Unified_Noise_Model()
unm.add_depolarizing_channel(depol_noise_prob,['rz'],[])
print(unm.noise_model)

#cost function

#There are the multipl
angles = []
for i in range(num_iterations):
    random_phi = randrange(0,20,step=1)
    random_phi = random_phi / 10
    angles.append(random_phi * pi)

print('angles:', angles)

result = minimize(
fun=cost_func2,
x0=angles,
args=(num_qubits,num_iterations,unm.noise_model),
method='L-BFGS-B',
)

print('result: ',result)
print('checking minimized parameters:')
cost = cost_func2(result.x,num_qubits,num_iterations,unm.noise_model)
print(cost)
```

```python
def test_results(optimized_operation_params,target_operation_params,gate_num=7,noise_model=None,debug = False):
    shots =10000
    qc_target = QuantumCircuit(1,1)
    qc_noisy = QuantumCircuit(1,1)
    qc_optimized = QuantumCircuit(1,1)

    #apply the target operation and the optimized operation
    for gate in range(gate_num):
            # apply one of the target gate
            qc_target.rz(target_operation_params[gate][0]*pi,0)
            qc_target.rx(pi/2,0)
            qc_target.rz(target_operation_params[gate][1]*pi,0)
            qc_target.rx(-pi/2,0)
            qc_target.rz(target_operation_params[gate][2]*pi,0)

            # apply one of the noisy gate
            qc_noisy.rz(optimized_operation_params[gate][0]*pi,0)
            qc_noisy.rx(pi/2,0)
            qc_noisy.rz(optimized_operation_params[gate][1]*pi,0)
            qc_noisy.rx(-pi/2,0)
            qc_noisy.rz(optimized_operation_params[gate][2]*pi,0)


            # apply one of the optimized gate
            qc_optimized.rz(optimized_operation_params[gate][0]*pi,0)
            qc_optimized.rx(pi/2,0)
            qc_optimized.rz(optimized_operation_params[gate][1]*pi,0)
            qc_optimized.rx(-pi/2,0)
            qc_optimized.rz(optimized_operation_params[gate][2]*pi,0)
    if debug:
          print('_____________________________________________________________________________________-')
          print('FUNCTION DEBUG')
          print('_____________________________________________________________________________________-')

          print('current optimized_operation_params: ',optimized_operation_params)
          print('noise model:',noise_model)

          print('target circuit:')
          display(qc_target.draw('mpl'))
          print('noisy circuit:')
          display(qc_noisy.draw('mpl'))
          print('optimized circuit:')
          display(qc_optimized.draw('mpl'))
          print('----------------------------------------------------------------------------------------')

    target_densitymatrix = DensityMatrix.from_instruction(qc_target)

    #-----GETTING NOISY AND OPTIMIZED DENSITY MATRIXES
    simulator = AerSimulator(noise_model=noise_model)
    
    noisy_tomography = StateTomography(qc_noisy)
    noisy_tomography_data = noisy_tomography.run(simulator,seed_simulation=100).block_for_results()
    noisy_densitymatrix=noisy_tomography_data.analysis_results('state').value
    
    optimized_tomography = StateTomography(qc_optimized)
    optimized_tomography_data = optimized_tomography.run(simulator,seed_simulation=100).block_for_results()
    optimized_densitymatrix=optimized_tomography_data.analysis_results('state').value
    #-----END - GETTING NOISY AND OPTIMIZED DENSITY MATRIXES -----

    noisy_fidelity = state_fidelity(target_densitymatrix,noisy_densitymatrix)
    optimized_fidelity = state_fidelity(target_densitymatrix,optimized_densitymatrix)
    return (noisy_fidelity,optimized_fidelity)
    
```

```python
from qiskit.quantum_info import state_fidelity,purity
from qiskit.circuit.library import HGate
from math import trunc
def cost_func3(optimized_operation_params,target_operation_params,noise_model=None,debug=False):
    if debug:
        print('_____________________________________________________________________________________-')
        print('FUNCTION DEBUG')
        print('_____________________________________________________________________________________-')

    
    # quantum circuit to apply the target operation and the optimized operation
    qc_target = QuantumCircuit(1)
    qc_optimized = QuantumCircuit(1)

    #set the initial state for the circuits. If None, then we start at |0>.
    global initial_state
    global next_initial_state
    
    if(initial_state != None):
        if debug: 
            print('Initial state:')
            display(initial_state.draw('bloch'))
        qc_target.initialize(initial_state.to_statevector())
        qc_optimized.initialize(initial_state.to_statevector())

    # get the amount of gates for the operations
    gate_num = len(target_operation_params) // 3


    #apply the target operation and the optimized operation
    for gate in range(gate_num):
            # apply one of the target gate
            qc_target.rz(target_operation_params[gate*3]*pi,0)
            qc_target.rx(pi/2,0)
            qc_target.rz(target_operation_params[gate*3 + 1]*pi,0)
            qc_target.rx(-pi/2,0)
            qc_target.rz(target_operation_params[gate*3 + 2]*pi,0)

            # apply one of the optimized gate
            qc_optimized.rz(optimized_operation_params[gate*3]*pi,0)
            qc_optimized.rx(pi/2,0)
            qc_optimized.rz(optimized_operation_params[gate*3 + 1]*pi,0)
            qc_optimized.rx(-pi/2,0)
            qc_optimized.rz(optimized_operation_params[gate*3 + 2]*pi,0)
    if debug:
        print('current optimized_operation_params: ',optimized_operation_params)
        print('noise model:',noise_model)

        print('target circuit:')
        display(qc_target.draw('mpl'))
        print('optimized circuit:')
        display(qc_optimized.draw('mpl'))
        print('----------------------------------------------------------------------------------------')

    # get statevector of quantum circuit after applying target operation
    target_statevector = DensityMatrix.from_instruction(qc_target)
    if debug:
        print('target operation:')
        display(target_statevector.draw('latex'))
        display(target_statevector.draw('bloch'))


    # get statevector of quantum circuit after applying optimized operation
    simulator = AerSimulator(noise_model=noise_model)

    optimized_tomography = StateTomography(qc_optimized)
    optimized_tomography_data = optimized_tomography.run(simulator,seed_simulation=100).block_for_results()
    optimized_statevector=optimized_tomography_data.analysis_results('state').value

    #Set the the initial state for the next optimization to be the current target_statevector
    next_initial_state = target_statevector

    if debug:
        print('optimized operation:')
        display(optimized_statevector.draw('latex'))
        #display(optimized_statevector.draw('bloch'))
        display(plot_bloch_multivector(optimized_statevector.data))
        print('----------------------------------------------------------------------------------------')

    cost = 1 - state_fidelity(target_statevector, optimized_statevector)
    if debug:
        print('purity of  of optimized operation:',purity(optimized_statevector).real)
        print('state fidelity:',state_fidelity(target_statevector, optimized_statevector))
        print('cost:',cost)
    return cost

initial_state = None
cost_func3([0,1/2,0],[0,1/2,0],[],debug=True)
initial_state = next_initial_state
cost_func3([0,1/2,0],[0,1/2,0],[],debug=True)
```

```python
def test_results(optimized_operation_params,target_operation_params,gate_num=7,noise_model=None,debug = False):
    shots =1024
    qc_noisy = QuantumCircuit(1,1)
    qc_optimized = QuantumCircuit(1,1)

    #apply the target operation and the optimized operation
    for gate in range(gate_num):
            # apply one of the noisy gate
            qc_noisy.rz(target_operation_params[gate][0]*pi,0)
            qc_noisy.rx(pi/2,0)
            qc_noisy.rz(target_operation_params[gate][1]*pi,0)
            qc_noisy.rx(-pi/2,0)
            qc_noisy.rz(target_operation_params[gate][2]*pi,0)


            # apply one of the optimized gate
            qc_optimized.rz(optimized_operation_params[gate][0]*pi,0)
            qc_optimized.rx(pi/2,0)
            qc_optimized.rz(optimized_operation_params[gate][1]*pi,0)
            qc_optimized.rx(-pi/2,0)
            qc_optimized.rz(optimized_operation_params[gate][2]*pi,0)
    qc_optimized.barrier()
    qc_optimized.compose(qc_noisy.inverse(),[0],inplace=True)
    qc_optimized.measure(0,0)
    qc_noisy.barrier()
    qc_noisy.compose(qc_noisy.inverse(),[0],inplace=True)
    qc_noisy.measure(0,0)


    if debug:
          print('_____________________________________________________________________________________-')
          print('FUNCTION DEBUG')
          print('_____________________________________________________________________________________-')

          #print('current optimized_operation_params: ',optimized_operation_params)
          #print('noise model:',noise_model)

          print('noisy circuit:')
          display(qc_noisy.draw('mpl'))
          print('optimized circuit:')
          display(qc_optimized.draw('mpl'))
          print('----------------------------------------------------------------------------------------')

    # calculate the noisy fideli and optimized fidelity
    noisy_simulator = AerSimulator(noise_model=noise_model)
  
    noisy_result = noisy_simulator.run(qc_noisy,shots=shots).result()
    noisy_counts = noisy_result.get_counts()
    noisy_fidelity = noisy_counts['0']/shots

    optimized_result = noisy_simulator.run(qc_optimized,shots=shots).result()
    optimized_counts = optimized_result.get_counts()
    optimized_fidelity =optimized_counts['0']/shots
    return (noisy_fidelity,optimized_fidelity)
    
```

```python
from qiskit.quantum_info import state_fidelity,purity
from qiskit.circuit.library import RZGate,RXGate
from math import trunc


#works only for one operation at the time
def cost_func4(optimized_operation_params,target_operation_params,noise_model=None,debug=False):
    global qc_reversed
    global qc
    global qc_final
    global update_circuit

    shots=1024

    if debug:
        print('_____________________________________________________________________________________-')
        print('FUNCTION DEBUG')
        print('_____________________________________________________________________________________-')


    qc_aux = qc.copy()
    qc_aux.rz(optimized_operation_params[0]*pi,0)
    qc_aux.rx(pi/2,0)
    qc_aux.rz(optimized_operation_params[1]*pi,0)
    qc_aux.rx(-pi/2,0)
    qc_aux.rz(optimized_operation_params[2]*pi,0)
    
    qc_aux.barrier()
    
    qc_aux.compose(RZGate(target_operation_params[2]*-pi,label="rz_ideal"),[0],inplace=True)
    qc_aux.compose(RXGate(pi/2,label="rx_ideal"),[0],inplace=True)
    qc_aux.compose(RZGate(target_operation_params[1]*-pi,label="rz_ideal"),[0],inplace=True)
    qc_aux.compose(RXGate(-pi/2,label="rx_ideal"),[0],inplace=True)
    qc_aux.compose(RZGate(target_operation_params[0]*-pi,label="rz_ideal"),[0],inplace=True)
    qc_aux.compose(qc_reversed,[0],inplace=True)
    
    qc_aux.measure([0],[0])
    # For other iterations of the same optimization, it is not necesary to update the circuit
    update_circuit = False

    if debug:
        print('current optimized_operation_params: ',optimized_operation_params)
        print('noise model:',noise_model)
        '''
        print('circuit (with optimized_operation_params):')
        display(qc.draw('mpl'))
        print('reversed circuit (with target_operation_params):')
        display(qc_reversed.draw('mpl'))
        print('final circuit for fidelity (qc+qc_reversed):')
        display(qc_final.draw('mpl'))'''
        display(qc_aux.draw('mpl'))
        print('----------------------------------------------------------------------------------------')

     #calculate the fidelity
    simulator = AerSimulator(noise_model=noise_model)
    result = simulator.run(qc_aux,shots=shots).result()
    counts = result.get_counts()

    #calculate the cost
    cost = 1 - counts['0']/shots
    if debug:
        print('COST: ',cost)
        #return fidelity as the cost
    return cost
```

```python
#Noise model and other params
depol_noise_prob = 0.000323
stateprep_noise_prob = 0
measurement_error_prob = 0.049
num_qubits = 1
num_gates_to_optimize = 30

#global circuits to modify
qc = QuantumCircuit(1,1)
qc_reversed = QuantumCircuit(1,1)
qc_final = QuantumCircuit(1,1)
update_circuit = True

unm = Unified_Noise_Model()
x_StatePreparation = XGate(label='x_StatePreparation')

#Cause there is no state preparation, this part of SPAM channel will not affect
unm.add_all_noise_channels(
state_preparation_error_prob=stateprep_noise_prob,
measurement_error_prob=measurement_error_prob,
state_preparation_error_gate=x_StatePreparation,
depolarizing_prob=depol_noise_prob,
one_qubit_gates=['rz','rx'],
one_qubit_gates_times=[[3.56E-8,3.56E-8],[3.56E-8,3.56E-8],[3.56E-8,3.56E-8],[3.56E-8,3.56E-8]],
two_qubits_gates=[],
two_qubits_gates_times=[],
qubits=range(num_qubits),
qubits_T1=[	0.0001,0.0001,0.0001,0.0001],
qubits_T2=[0.0001,0.0001,0.0001,0.0001]
)

results = []
all_target_operation_params = []
for current_gate in range(num_gates_to_optimize):
    print('GATE NUMBER ',current_gate)
    #Setting up the angles of the decomposed gate
    #for our first trial, we will just give angles for U(-pi/2,pi/2,0)
    random_params = [
        randrange(0,21)/10,
        randrange(0,21)/10,
        randrange(0,21)/10
    ]

    #display(DensityMatrix.from_instruction(qc).draw('bloch'))
    target_operation_params = []
    target_operation_params.append(random_params[0])
    target_operation_params.append(random_params[1])
    target_operation_params.append(random_params[2])

    #Setting up the random angles to optimize the decomposed gate
    optimized_operation_params = []
    optimized_operation_params.append(random_params[0])
    optimized_operation_params.append(random_params[1])
    optimized_operation_params.append(random_params[2])

    print('optimized operation params:', optimized_operation_params)
    print('target operation params:',target_operation_params)

    #if current_gate % 7 == 0:
    all_target_operation_params.append(random_params)
    
    result=minimize(
    fun=cost_func4,
    x0=optimized_operation_params,
    args=(target_operation_params,unm.noise_model,False)
    ,method='L-BFGS-B',
    options={
    #'eps':0.5,
    #'ftol':0.0000001
    })
    results.append(result)
    
    print('RESULT: ',result)
    if False:
        print('circuit (with optimized_operation_params):')
        display(qc.draw('mpl'))
        print('reversed circuit (with target_operation_params):')
        display(qc_reversed.draw('mpl'))
        print('final circuit for fidelity (qc+qc_reversed):')
        display(qc_final.draw('mpl'))
        
    #update the global circuits
    if True:
        qc.rz(result.x[0]*pi,0)
        qc.rx(pi/2,0)
        qc.rz(result.x[1]*pi,0)
        qc.rx(-pi/2,0)
        qc.rz(result.x[2]*pi,0)
        qc.barrier()

        qc_aux = QuantumCircuit(1,1)
        qc_aux.compose(RZGate(target_operation_params[2]*-pi,label="rz_ideal"),[0],inplace=True)
        qc_aux.compose(RXGate(pi/2,label="rx_ideal"),[0],inplace=True)
        qc_aux.compose(RZGate(target_operation_params[1]*-pi,label="rz_ideal"),[0],inplace=True)
        qc_aux.compose(RXGate(-pi/2,label="rx_ideal"),[0],inplace=True)
        qc_aux.compose(RZGate(target_operation_params[0]*-pi,label="rz_ideal"),[0],inplace=True)
        qc_aux.compose(qc_reversed,[0],inplace=True)
        qc_reversed = qc_aux
        
        #qc_final is the qc + qc_reversed circuit to calculate the fidelity
        qc_final = qc.compose(qc_reversed,[0])
        qc_final.measure([0],[0])
    update_circuit = True
#time?
```
